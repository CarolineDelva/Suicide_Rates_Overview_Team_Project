```python

import pandas as pd
import numpy as numpy

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine 

from flask import Flask, jsonify, render_template
from flask_sqlalchemy import SQLAlchemy

import requests 
import json
from config import api_key
from config import google_api_key

import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import os
```

```python
POSTGRES = {
'user': 'postgres',
'pw': api_key,
'db': 'suicide_rates_overview_db',
'host': 'localhost',
'port': '5432',
}

engine = create_engine('postgresql://%(user)s:%(pw)s@%(host)s:%(port)s/%(db)s'% POSTGRES)
df_suicide_rates_df = pd.read_sql('select * from suicide_rates_db', con=engine)
df_suicide_rates_df
```

```python
bins = [0, 4467, 8934, 13401, 17868, 22338 ]
names = ["low", "mid-low", "mid", "mid-high", "high"]
d = dict(enumerate(names, 1))
```

```python
df_suicide_rates_df['suicides_no'] = np.vectorize(d.get)(np.digitize(df_suicide_rates_df['suicides_no'], bins))
df_suicide_rates_df
```

```python
#Assign the data to X and y
from sklearn.model_selection import train_test_split
X = df_suicide_rates_df[["year", "population", "suicideshundredk", "gdp_for_year", "gdp_per_capita"]]
y = df_suicide_rates_df["suicides_no"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler 
X_scaler = MinMaxScaler().fit(X_train) 
X_train_scaled = X_scaler.transform(X_train) 
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_train
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC
model2 = SVC(kernel='linear')
model2.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

```python
plt.scatter(model2.predict(X_train), model2.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model2.predict(X_test), model2.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
### END SOLUTION
```
